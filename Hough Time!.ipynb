```python
def getPos(theta, Dec, RA):
    return -Dec*np.cos(theta[::-1]*(np.pi/180)) - RA*np.sin(theta[::-1]*(np.pi/180))
```

```python
def bin2angle(b):
    return b * (tmax - tmin) / theta.shape[0]

def angle2bin(t):
    return t / (tmax - tmin) * theta.shape[0]

def bin2pos(b):
    return (b - (posa.shape[0] / 2)) * (pmax - pmin) / posa.shape[0]

def pos2bin(p):
    return p / (pmax - pmin) * posa.shape[0] + (posa.shape[0] / 2)
```
